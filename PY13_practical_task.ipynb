# 9.1. Соотношение ушедших и лояльных клиентов

- клиенты накопили нужную сумму для определенной покупки и ушли,
- переложили денежные средства в другой банк под более выгодный процент,

# 9.10. Влияние кредитного рейтинга и срок пользования услугами банка на отток клиентов

Переведите числовой признак CreditScore в категориальный. Для этого воспользуйтесь функцией get_credit_score_cat(), которая приведена ниже. Примените её к столбцу CreditScore и создайте новый признак CreditScoreCat — категории кредитного рейтинга.

Постройте сводную таблицу, строками которой являются категории кредитного рейтинга (CreditScoreCat), а столбцами — количество лет, в течение которых клиент пользуется услугами банка (Tenure). В ячейках сводной таблицы должно находиться среднее по признаку оттока (Exited) — доля ушедших пользователей.

На основе полученной сводной таблицы постройте тепловую карту с аннотацией. Найдите на тепловой карте категории клиентов, которые уходят чаще всего.

```python
def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score >= 851:
        return "Top"
    elif credit_score < 300:
        return "Deep"

churn_data['CreditScoreCat'] = churn_data['CreditScore'].apply(get_credit_score_cat)

pivot = churn_data.pivot_table(
     values='Exited',
     index='CreditScoreCat',
     columns='Tenure'
 )
fig = plt.figure(figsize=(15, 8))

heatmap = sns.heatmap(data=pivot, cmap='plasma', annot=True)
heatmap.set_title(
    'Влияние кредитного рейтинга и срока пользования услугами банка на отток клиентов', 
    fontsize=16
        );
heatmap.set_xlabel('Количество лет пользования услугами банка', fontsize=14)
heatmap.set_ylabel('Категории кредитного рейтинга', fontsize=14);
```

На тепловой карте видно, что самый большой отток клиентов среди тех, кто пользовался услугами банка менее года и имеет низкий кредитный рейтинг (от 300 до 499). На втором месте клиенты с таким же рейтингом, но пользующиеся услугами банка 10 лет.

Можно предположить, что клиенты, давно пользующиеся услугами банка, покидают банк, т.к. отпадает необходимость в дальнейших услугах банка. А новых клиентов не устраивают условия банка.

Таким образом, сильно маленький или сильной большой срок пользования услугами банка влияет на отток клиентов, а также низкий кредитный рейтинг, что вполне логично, т.к. у клиентов с низким уровнем кредитного рейтинга отсутствут возможность выплачивать кредит.
